```python
x=iris.iloc[:,:4]# For x i gave First three columns Because it has  variables
y=iris.iloc[:,-1]#for y, i gave last colomn
```

```python
x # checking x
```

```python
y #checking y
```

     Data normalization

fit_transform is from SKLEARN
Used on the training data. Use it to extract Parameter from training data.

```python
x=preprocessing.StandardScaler().fit_transform(x)
```

```python
x #The values has been changed. Our first value was different and Current value is a range using data normalization
```

# training and tesing the modle
80% training
20% tesing

```python
x_train,x_test,y_train,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train.shape # Out of 150, 120 is kept for training.
```

```python
x_test.shape # Out of 150, 30 is kept for testing.
```

# MODEL AND PREDICTED

```python
knnmodel=KNeighborsClassifier(n_neighbors=3)
```

```python
knnmodel.fit(x_train,y_train)# just made a mode by the splitted data we did
```

```python
y_predict=knnmodel.predict(x_test)#we will put x_test as an input ad ill get y as a output . y predicted
```

# **Accuracy Score**

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(ytest.values,y_predict)
```
